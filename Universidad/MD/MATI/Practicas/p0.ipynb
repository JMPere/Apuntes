<h3>Representando grafos</h3>
<p>Los siguientes ejemplos muestran cómo cambiar el color de los vértices y su tamaño o el de las aristas. Para ello se crea un diccionario (una lista entre llaves) de modo que cada entrada de la lista sea una pareja separada por :, el elemento de la izquierda será el color y el de la derecha la lista de elementos a dibujar con ese color.</p>

```python
P = graphs.PetersenGraph()
d = {'#FF0000':[0,5], '#FF9900':[1,6], '#FFFF00':[2,7], '#00FF00':[3,8], '#0000FF':[4,9]}
P.plot(vertex_colors=d,vertex_size=500).show(figsize=4)
```

<p>¿Cómo quedará este grafo?. Dada una cadena de texto, se genera un grafo que une cada letra con la siguiente en el texto. Se incluyen opciones gráficas para etiquetar los vértices con las propias letras y se etiquetan y colorean las aristas de forma que todas las aristas con la misma etiqueta llevan el mismo color:</p>

```python
texto = 'estoesunaprueba'
g = Graph({}, loops=True, multiedges=True)
m=0
for i,j in [(texto[k], texto[k+1]) for k in range(len(texto)-1)]:
    m=m+1
    g.add_edge(i, j, mod(m,3))
g.plot(color_by_label=True,edge_labels=True, edge_style='solid').show(figsize=(7))
```

<h3>Grafo dirigido y/o etiquetado</h3>
<p>Todo lo anterior es válido para grafos dirigidos, sin más que cambiar <strong><em>Graph</em></strong> por <strong><em>DiGraph</em></strong>.</p>

```python
texto = 'estoesunaprueba'
g = DiGraph({}, loops=True, multiedges=True)
m=0
for i,j in [(texto[k], texto[k+1]) for k in range(len(texto)-1)]:
    m=m+1
    g.add_edge(i, j, mod(m,3))
g.plot(color_by_label=True,edge_labels=True, edge_style='solid').show(figsize=(7))
```

```python
Creamos los conjuntos S y T, compuestos de los enteros de 0 a 102.
Buscaremos que siendo i,j un entero del conjunto S, recogemos el elemento de la lista que siendo
|i-j| no sea cuadrática, y lo contrario para el conjunto T.
```

```python
C=[] #Conjunto [0.....x]
S=[] #Conjunto S
for a in range(50+53):
    C.append(a)
S.append(C[0])
valido=false
for i in range(1,len(C)-1):
    valido=true
    #print(i)
    for j in S:
        #print('\n|',i,'-',j,'|',j,' es ',sqrt(abs(i-j)), ", es entero: ",sqrt(abs(i-j)).is_integer())
        if sqrt(abs(i-j)).is_integer() == true :
            valido=false
    if valido :
        S.append(i)
        
print(S)
```
